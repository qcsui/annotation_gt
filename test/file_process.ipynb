## First screening of the files

```python
import os

def read_file_list(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return {line.strip() for line in file}


def convert_filename(filename):
    base_name = filename.split('_p')[0]  # 获取文件名的基础部分
    new_filename = f"{base_name}.pdf"  # 拼接新的文件名
    return new_filename

def clean_directory(target_folder, file_list):
    for filename in os.listdir(target_folder):
        file_path = os.path.join(target_folder, filename)
        if os.path.isfile(file_path) and filename not in file_list:
            os.remove(file_path)
            print(f'Removed: {filename}')
        else:
            print(f'Kept: {filename}')
```

```python

html_file_list_path = '..\\app\\static\\checked_files_0.txt'
html_target_folder = '..\\app\\static\\data\\htmls'
pdf_target_folder = '..\\app\\static\\data\\pdfs'
file_list_html = read_file_list(html_file_list_path) # 读取html文件列表
clean_directory(html_target_folder, file_list_html) # 清理html文件夹
file_list_pdf = [convert_filename(html_filename) for html_filename in file_list_html] # 转换html文件名为pdf文件名
clean_directory(pdf_target_folder, file_list_pdf) # 清理pdf文件夹

```
